```sql
--- DATA EXPLORATION
--- Preview the tables
SELECT *
FROM spend_analysis_dataset
LIMIT 10;

--- new columns
SELECT *,
       YEAR(PurchaseDate) AS Year,
       MONTH(PurchaseDate) AS Month
FROM spend_analysis_dataset;
-------------------------------------------------
--- Working with transformed Dataset
-- Total Spend
SELECT SUM(TotalCost) AS Total_Spend 
FROM spend_analysis;

-- Spend by Supplier
SELECT Supplier, SUM(TotalCost) AS Total_Spend
FROM spend_analysis
GROUP BY Supplier
ORDER BY Total_Spend DESC;

-- Spend by Category
SELECT Category, SUM(TotalCost) AS Total_Spend
FROM spend_analysis
GROUP BY Category
ORDER BY Total_Spend DESC;

-- Monthly Spend
SELECT Month, SUM(TotalCost) AS Monthly_Spend
FROM spend_analysis
GROUP BY Month
ORDER BY Month;

-- Top Buyers
SELECT Buyer, SUM(TotalCost) AS Total_Spend
FROM spend_analysis
GROUP BY Buyer
ORDER BY Total_Spend DESC;

--- Most Purchased Items
SELECT ItemName, SUM(Quantity) AS Total_Quantity
FROM spend_analysis
GROUP BY ItemName
ORDER BY Total_Quantity DESC;

---Average Unit Price per Category
SELECT Category, AVG(UnitPrice) AS Avg_Unit_Price
FROM spend_analysis
GROUP BY Category;

--- Supplier Transaction Count
SELECT Supplier, COUNT(*) AS Transactions
FROM spend_analysis
GROUP BY Supplier
ORDER BY Transactions DESC;
```
